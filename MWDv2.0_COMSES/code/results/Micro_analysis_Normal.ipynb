```python
#for Data Frame
import pandas as pd
#For plotting
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_micro= pd.read_csv('series_Normal_micro_1000ts_1reps_2023-03-09 16:47:40.csv')
```

```python
pd.options.display.max_columns = None
display(df_micro)
```

```python
fig = sns.relplot(data=df_micro, x="step", y='alpha' , 
            kind="line", hue='household')
fig.set(ylabel=r'$alpha$', title=r'Distribution of $\alpha$ for the households')

plt.show()
```

```python
#For filter the alpha
workaholics = df_micro[df_micro['alpha'] < 0.5]

fig = sns.relplot(data=df_micro, x="step", y='H_O' , 
            kind="line", hue='household')
fig.set(ylabel=r'$H_O$', title='Discretionary hours')

plt.show()
```

```python
#Wage that firm pays to households 
#For filter the time step 0
#h1 = df_micro[df_micro['step'] > 0]

fig = sns.relplot(data=df_micro, x="step", y='omega' , 
            kind="line", hue='household')
fig.set(ylabel='wage')

plt.show()
```

```python
#income housholds = w(H_0 + H_N)
#For filter the time step 0
h1 = df_micro[df_micro['step'] > 0]

fig = sns.relplot(data=h1, x="step", y='income' , 
            kind="line", hue='household')
fig.set(ylabel='income')

plt.show()
```

```python
fig = sns.relplot(data=df_micro, x="step", y='m' , 
            kind="line", hue='household')
fig.set(ylabel='monetary holdings')

plt.show()
```
